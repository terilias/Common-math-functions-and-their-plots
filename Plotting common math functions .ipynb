```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
# This cell is here only for the development purposes and not for the learning aim of this notebook.

def to_latex(params):
    """
    Prepare the TeX markup to print a math function with many different parameters.
  
    This function takes a list of strings and combine them together inside '$' symbols to make a TeX compatible
    string. Returns this string so can be used to plt.title() for the graphs of this notebook.
  
    Parameter:
    params (list): The math function's parameters to include to the TeX string (e.g. ['x',^{',str(z),'}','+','c']). 
  
    Returns:
    to_print: A string that start with '$f(x)=' contains the contents of params list and ends with the '$' symbol. 
              
    """
    to_print = '$f(x)='
    for param in params:
        to_print = to_print + param
    to_print = to_print + '$'
    return to_print
```

##  1) $ f(x)=x^z,\ z \in \mathbb{z}$


In the first plot $z$ is an even number. We state this using the following math expression:  $f(x)=x^z,\ z=2k,\ k\in \mathbb{Z}$ <br>
Υou can experiment in the cell below, changing the $z$ value to understand the behavior of this math function. 
If you want to plot the $f(x) = x$ line assigne the value $1$ to $z$ variable and if you assigne it the $0$ value you will have $f(x) = 1$ because any number raised to $0$ it gives $1$.

```python
# Creating vectors X and Y to plot this very simple function.
x = np.linspace(-5, 5, num=100)
z = 2 # Initially is z = 2
if z % 2 != 0:
    print("Warning: In this function you are using an odd number, not an even.\nIf you want to see the plot for the function with an even pοwer, change z's value inside the code on the above cell.")
y = x ** z

# Prepare the plot.
plt.xlabel('X axis')
plt.ylabel('Y axis')
# Note that we can use a subset TeX markup in any matplotlib text string
# by placing it inside a pair of dollar signs ($), as we already do in the Markdown sections of this notebook.
plt.title(to_latex(['x','^{',str(z),'}']))
#plt.title('$f(x)=x^{'+str(z)+'}$') 

# Create and show the plot. Without using .show() some undesirable informations are shown.
plt.plot(x,y)
plt.show()
```

## 2) $f(x)=\alpha x^z,\alpha \in \mathbb{R}$
Now, let's try to add the parameter $\alpha$ to our function.
Try to set different values to parameter a (e.g. 1, 2, 8) to see that $\alpha$ indicates how "fat" or how "skinny" the parabola will be.

```python
a = 0.08 # Initially is a = 0.08
z = 2 # Initially is z = 2
y = a * x**z
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)='+str(a)+'x^{'+str(z)+'}$')
plt.xlim([-10,10])
plt.ylim([0,2])
plt.plot(x,y)
plt.show()
```

And as you may guess if we set a negative value in parameter $\alpha$ we will have a vertically mirrored graph.

```python
a = -0.08 # Initially is a = -0.08
z = 2 # Initially is z = 2
y = a * x**z
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)='+str(a)+'x^{'+str(z)+'}$')
plt.xlim([-10,10])
plt.ylim([-2,2])
plt.plot(x,y)
plt.show()
```

## 3) $f(x)=\alpha x^z+\beta,\ \ \ \alpha, \beta, z \in \mathbb{R}$

We are nw going to add one more parameter to our function, the parameter $\beta$ wich is called intercept and defines the point where our graph meets the Y axis.

Try to set different values to $\beta$ to see the impact of this parameter to our graph. *Tip: use 4, 0, -0.5, -1, -4 with this order.* 

```python
a = 2 # Initially is a = 2
z = 2 # Initially is z = 2
b = 1 # Initially is b = 1 
y = a * x**z + b
plt.title('$f(x)='+str(a)+'x^{'+str(z)+'}+'+str(b)+'$')
# How to centralize the Y axis: https://www.geeksforgeeks.org/how-to-put-the-origin-in-the-center-of-the-figure-with-matplotlib/
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero') 
plt.xlim([-5,5])
plt.ylim([-1,5])
plt.plot(x,y)
plt.show()
```

In our case that it's shown below, we have $n = 4$, so we are going to plot this function: $f(x)=\alpha_4 x^4+\alpha_{3} x^{3} + \alpha_2 x^2 + a_1x + a_0$.
Try to change the parameters values and you will see different shapes! For example try to reduce the polynomial's degree by assigning $0$ to $a_4$. 

*Machine Learning Usecase: Polynomial Regression with this form:* $f(x)=b_0 + b_1x + b_2x^2 + ... + b_nx^n$

## 5) $f(x)=\alpha \cdot x,\ \alpha \in \mathbb{R}$
Η βασική εξίσωση μιας ευθείας όπου το $\alpha$ είναι ο συντελεστής διεύθυνσης (ή κλίση - slope ή gradient στα αγγλικά) της ευθείας και ισούται με την εφαπτομένη της γωνίας που σχηματίζει η ευθεία με τον άξονα x'x.

Επειδή όπως είπαμε ήδη το $\alpha$ είναι ίσο με την εφαπτομένη της γωνίας, συμπαιρένουμε ότι αν το $\alpha$, δηλαδή ο συντελεστής διεύθυνσης της ευθείας είναι θετικός, αν η γωνία που σχηματίζει με τον άξονα x'x είναι οξεία...

```python
a = 1.8 # Initially is a = 1.8
y = a * x
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)=' + str(a) + '*' + 'x' + '$')

plt.xlim([0,5])
plt.ylim([0,5])
plt.show()
```

...και αρνητικός, αν είναι αμβλεία:

```python
a = -1.8 # Initially is a = -1.8
y = a * x
plt.plot(x,y)
plt.title('$f(x)=' + str(a) + '*' + 'x' + '$')
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero') 
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()
```

Το πρόσημο της κλίσης δηλαδή μπορούμε να το βρούμε κοιτώντας απλά την γραφική παράσταση της ευθείας. 

*Σημείωση πρακτικής εφαρμογής της έννοιας στους αλγορίθμους μηχανικής μάθησης*:
Αυτό είναι ιδιαίτερα χρήσιμο στον πασίγνωστο αλγόριθμο βελτιστοποίσης *επικλινής κάθοδος, Gradient Descent* επειδή με την κλίση της εφαπτομένης της συνάρτησης κόστους (δηλαδή με την κλίση της ευθείας εκείνης που εφάπτεται στην γραφική παράσταση της συνάρτησης κόστους) υπολογίζουμε το πώς πρέπει ακριβώς να μεταβάλλουμε τις παραμέτρους της συνάρτησης για να βρούμε το ελάχιστό της. Εάν η κλίση της εφαπτομένης είναι αρνητική ("γέρνει" δηλαδή προς τα αριστερά όπως αυτή που βλέπουμε παραπάνω) τότε σύμφωνα με τον αλγόριθμο αυξάνουμε την τιμή των παραμέτρων (άρα πάμε προς τα δεξιά) γιατί στην πράξη σημαίνει ότι "πάμε προς την σωστή κατεύθυνση", δηλαδή προς τον πάτο του "μπολ" της γραφικής παράστασης της συνάρτησης κόστους. (Το ότι αυξάνουμε την τιμή εξηγείται αλγεβρικά από τα πρόσημα: το μείον της εξίσωσης με την οποία ενημερώνουμε τις παραμέτρους, αναιρεί το μείον της αρνητικής κλίσης άρα έρχεται το συν, δηλαδή αυξάνεται η τιμή της παραμέτρου). Εάν η κλίση είναι θετική ("γέρνει" δηλαδή προς τα δεξιά) τότε μειώνουμε τις τιμές των παραμέτρων (πάμε προς τα αριστερά, και το μείον της εξίσωσης ενημέρωσης των τιμών των παραμέτρων δεν αναιρείται) και πλησιάζουμε το ελάχιστο, τον πάτο του μπολ. 
Εκτός από το πρόσημο βέβαια, βρίσκοντας την τιμή της κλίσης (πάντα με μερική παράγωγο) εντοπίζουμε και το μέγεθος κατά το οποίο θα μειώσουμε (ή θα αυξήσουμε αν είμαστε στην αντίστοιχη περίπτωση) τις τιμές των παραμέτρων. 

Βρίσκοντας το ελάχιστο (minimization) της συνάρτησης κόστους, έχουμε βρει τις κατάλληλες τιμές των παραμέτρων που οδηγούν στο ελάχιστο, άρα έχουμε βρει την βέλτιστη συνάρτηση-μοντέλο.

## 6) $f(x)=\alpha \cdot x + \beta ,\ \alpha, \beta  \in \mathbb{R}$
Η βασική γραμμική εξίσωση, με μια ακόμα παράμετρo, την $\beta$ η οποία ορίζει το σημείο που η ευθεία τέμνει τον άξονα Y. 

Δοκιμάστε να αλλάξετε την τιμή της παραμέτρου $\beta$ για να δείτε πως η ευθεία (η γραμμή στην προκειμένη περίπτωση) αλλάζει θέση (π.χ. δώστε για τιμή το 0, το 3, το -2 κά). Στους αλγορίθμους Μηχανικής Μάθησης, η παράμετρος $\beta$ έχει τον ρόλο του bias και υποδυκνύει το  κατώφλι (threshold) στα μοντέλα ταξινόμησης ενώ στα μοντέλα παλινδρόμησης λέγεται intercept. 

```python
a = 0.8 # Initially is a = 0.8
b = 0.6 # Initially is b = 0.6
y = b + a * x
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)=' + str(b) + '+' + str(a) + '*' + 'x' + '$')
plt.xlim([0,5])
plt.ylim([-3,5])
plt.show()
```

## 7) Circle, $x^2 + y^2 = 1$
Στην συνέχεια σχεδιάζουμε τον μοναδιαίο κύκλο, δηλαδή έναν κύκλο με ακτίνα ίση με 1 χρησιμοποιώντας τις παραμετρικές του εξισώσεις, καθώς δεν υπάρχει συνάρτηση που η γραφική της παράσταση να είναι ένας κύκλος.

## 8) Sine, $f(x) = sin(x)$
Ως συνάρτηση το ημίτονο είναι περιοδική με περίοδο Τ=2$π$ και περιττή.Το ότι έχει περίοδο Τ=2$π$ σημαίνει ότι για $x=0, 2π, 4π, 6π,...$ κάθε φορά που προσθέτουμε δηλαδή 2$π$ στην είσοδο, τότε η έξοδος είναι η ίδια. Δείτε το αυτό στο σχήμα παρακάτω. Το κύμα που βλέπουμε να ξεκινάει εκεί που $x=0$ ολοκληρώνεται εκεί που το x είναι περίπου 6,28 -δηλαδή 2$π$- και επαναλαμβάνεται στη συνέχεια (και αριστέρα και δεξιά στον οριζόντιο άξονα). Μία συνάρτηση f(x) με πεδίο ορισμού το Af λέγεται άρτια, αν για κάθε x που ανήκει στο Af ισχύει ότι το -x ανήκει στο Af και ότι f(-x)=f(x). (Πεδίο ορισμού είναι το σύνολο των τιμών που μπορούν να δοθούν ως είσοδος σε μία συνάρτηση). Συνήθως χρησιμοποιούμε το δεύτερο ορισμό με πεδίο ορισμού το σύνολο των πραγματικών αριθμών. Το σύνολο τιμών της ημιτονοειδούς συνάρτησης είναι το [-1,1].Το βλέπουμε αυτό στο γράφημα με το ότι οι τιμές στον άξονα Y είναι πάντα ανάμεσα στο -1 και το 1.

```python
x = np.linspace(-20, 20, num=150)
y = np.sin(x)
plt.plot(x,y)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero') 
plt.title(to_latex(['sin(x)']))
plt.xlim([-20,20])
plt.ylim([-2,2])
plt.show()
```

## 9) Cosine, $f(x) = cos(x)$
Ως συνάρτηση το συνημίτονο είναι περιοδική με περίοδο Τ=2$π$ και άρτια. Μία συνάρτηση f(x) με πεδίο ορισμού το Af λέγεται άρτια, αν για κάθε x που ανήκει στο Af ισχύει ότι το -x ανήκει στο Af και ότι f(-x)=f(x). Παρατηρείστε το στο γράφημα παρακάτω. Συνήθως χρησιμοποιούμε το δεύτερο ορισμό με πεδίο ορισμού το σύνολο των πραγματικών αριθμών. ο σύνολο τιμών της εκθετικής συνάρτησης είναι το [-1,1] όπως άλλωστε και η συνάρτηση ημιτόνου.

```python
y = np.cos(x)
plt.plot(x,y)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero') 
plt.title(to_latex(['cos(x)']))
plt.xlim([-20,20])
plt.ylim([-2,2])
plt.show()
```
