##  1) $ f(x)=x^z,\ z \in \mathbb{z}$


In the first plot $z$ is an even number. We state this using the following math expression:  $f(x)=x^z,\ z=2k,\ k\in \mathbb{Z}$ <br>
Î¥ou can experiment in the cell below, changing the $z$ value to understand the behavior of this math function. 
If you want to plot the $f(x) = x$ line assigne the value $1$ to $z$ variable and if you assigne it the $0$ value you will have $f(x) = 1$ because any number raised to $0$ it gives $1$.

```python
# Creating vectors X and Y to plot this very simple function.
x = np.linspace(-5, 5, num=100)
z = 2 # Initially is z = 2
if z % 2 != 0:
    print("Warning: In this function you are using an odd number, not an even.\nIf you want to see the plot for the function with an even pÎ¿wer, change z's value inside the code on the above cell.")
y = x ** z

# Prepare the plot.
plt.xlabel('X axis')
plt.ylabel('Y axis')
# Note that we can use a subset TeX markup in any matplotlib text string
# by placing it inside a pair of dollar signs ($), as we already do in the Markdown sections of this notebook.
plt.title('$f(x)=x^{'+str(z)+'}$') 

# Create and show the plot. Without using .show() some undesirable informations are shown.
plt.plot(x,y)
plt.show()
```

### 
In the second plot we have the same simple function but here the power is an odd number:
$f(x)=x^z,\ z=2k+1,\ k\in \mathbb{Z}$ <br>
Replace the value of $z$ with bigger or lower odd numbers to see how the plot changes.

```python
z = 3 # Initially is z = 3
if z % 2 == 0:
    print("Warning: In this function you are using an even number, not an odd.\nIf you want to see the plot for the function with an odd pÎ¿wer, change z's value inside the code on the above cell.")
y = x ** z
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)=x^'+str(z)+'$')
plt.plot(x,y)
plt.show()
```

## 2) $f(x)=\alpha x^z,\alpha \in \mathbb{R}$
Now, let's try to add the parameter $\alpha$ to our function.
Try to set different values to parameter a (e.g. 1, 2, 8) to see that $\alpha$ indicates how "fat" or how "skinny" the parabola will be.

```python
a = 0.08 # Initially is a = 0.08
z = 2 # Initially is z = 2
y = a * x**z
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)='+str(a)+'x^{'+str(z)+'}$')
plt.xlim([-10,10])
plt.ylim([0,2])
plt.plot(x,y)
plt.show()
```

And as you may guess if we set a negative value in parameter $\alpha$ we will have a vertically mirrored graph.

```python
a = -0.08 # Initially is a = -0.08
z = 2 # Initially is z = 2
y = a * x**z
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('$f(x)='+str(a)+'x^{'+str(z)+'}$')
plt.xlim([-10,10])
plt.ylim([-2,2])
plt.plot(x,y)
plt.show()
```

Before move to the next graph try to set an odd number as a value for the parameter z and after that change the sign of $\alpha$ to see how to get an horizontally mirrored graph.

## 3) $f(x)=\alpha x^z+\beta,\ \ \ \alpha, \beta, z \in \mathbb{R}$

We are nw going to add one more parameter to our function, the parameter $\beta$ wich is called intercept and defines the point where our graph meets the Y axis.

Try to set different values to $\beta$ to see the impact of this parameter to our graph. *Tip: use 4, 0, -0.5, -1, -4 with this order.* 

```python
a = 2 # Initially is a = 2
z = 2 # Initially is z = 2
b = 1 # Initially is b = 1 
y = a * x**z + b
plt.title('$f(x)='+str(a)+'x^{'+str(z)+'}+'+str(b)+'$')
# How to centralize the Y axis: https://www.geeksforgeeks.org/how-to-put-the-origin-in-the-center-of-the-figure-with-matplotlib/
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero') 
plt.xlim([-5,5])
plt.ylim([-1,5])
plt.plot(x,y)
plt.show()
```

## 4) Polynomial,  $f(x)=\alpha_n x^n+\alpha_{n-1} x^{n-1} + ... + \alpha_2 x^2 + a_1x + a_0,\ \ \ \alpha_{i,\  i \in [0,n]}\in \mathbb{R}, \ n \in \mathbb{N}$
In our case that it's shown below, we have $n = 4$, so we are going to plot this function: $f(x)=\alpha_4 x^4+\alpha_{3} x^{3} + \alpha_2 x^2 + a_1x + a_0$.
Try to change the parameters values and you will see different shapes! For example try to reduce the polynomial's degree by assigning $0$ to $a_4$. 

*Machine Learning Usecase: Polynomial Regression with this form:* $f(x)=b_0 + b_1x + b_2x^2 + ... + b_nx^n$

```python
# I wanted to plot this but I could not :Î¡ https://commons.wikimedia.org/wiki/File:Polynomialdeg3.svg

a4 = 0.05 # Initially is a4 = 0.5
a3 = 0.25 # Initially is a3 = 0.25
a2 = 0.75 # Initially is a2 = 0.75
a1 = 1.5 # Initially is a1 = 1.5
a0 = 2 # Initially is a0 = 2

n = 4 # Initially is n = 4

y = a4*x**n + a3 * x**(n-1) + a2 * x**(n-2)  + a1*x + a0

plt.title('$f(x)='+str(a4)+'x^{'+str(n)+'}+'+str(a3)+'x^{'+str(n-1)+'}+'+str(a2)+'x^{'+ str(n-2)+'}+'+str(a1)+'x'+'+'+str(a0)+'$')

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero') 

plt.xlim([-5,3])
plt.ylim([-20,20])
plt.plot(x,y)
plt.show()
```

Now, we leave behind us the curved lines to study the straight lines.
